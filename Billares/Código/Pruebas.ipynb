```python
import numpy as np
import matplotlib.pyplot as plt


eigs = []

a = 2
b = 3

for n1 in range(200):
    for n2 in range(200):
        eigs.append(np.pi*(n1**2/a**1 + n2**2/b**2))
```

```python
eigs.sort()
```

```python
eigs
```

```python
sep = []
for i in range(len(eigs)-1):
    sep.append(eigs[i+1]-eigs[i])
    
sep = sep/np.mean(sep)
```

```python
sep
```

```python
plt.hist(sep,100,[0,20],density=True)
x = np.linspace(0,20,100)
plt.plot(x,np.exp(-x))
plt.show()
```

```python
s = np.linspace(0,4,100)
def g(s):
    gg = np.pi/2*s*np.exp(-np.pi/4*s**2)
    return(gg)

def pois(s):
    gg = np.exp(-s)
    return(gg)

gs = [g(ss) for ss in s ]
gpois = [pois(ss) for ss in s]

plt.legend()

plt.plot(s,gs,label="GOE")
plt.plot(s,gpois,label="Poisson")
```
