```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = np.linspace(0,2*np.pi, 256).reshape(256, 1) # the angle of the projection in the xy-plane

zet = np.linspace(0, 2*np.pi, 256).reshape(-1, 256) # the angle from the polar axis, ie the polar angle

# Transformation formulae for a spherical coordinate system.
x = 2*np.cos(phi)
y = 1*np.sin(phi)
z = zet

fig = plt.figure()  # Square figure
ax = fig.add_subplot(111, projection='3d')

ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))
ax.plot_surface(x, y, z, color='b',alpha=0.5)

ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$',fontsize=20)

plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
ax.set_zticks([0, np.pi, 2*np.pi])
ax.set_zticklabels(['0','','2'])

ax.set_zlabel('', fontsize=20, rotation = 0)

plt.show()
    
```
